```python
import joblib

# 1. Load the model
model = joblib.load("tone_model.pkl")
```

```python
# 2. Test with a few samples
examples = [
    "The new feature works perfectly and saves a lot of time.",
    "The new feature works perfectly and saves a lot of time.",
    "The experience was okay, nothing special."
]
```

```python
predictions = model.predict(examples)

for text, tone in zip(examples, predictions):
    print(f"{tone.upper():<10} | {text}")
```
