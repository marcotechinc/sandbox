```python
import pandas as pd
from tqdm import tqdm

df = pd.read_csv("feeds_rows.csv")

def synthesize(row):
    if pd.notna(row["content"]):
        return row["content"]
    return (
        f"This article, titled '{row.title}', explores recent developments related to {row.topic.lower()}. "
        f"Published on {row.source}, it falls under the {row.category.lower()} category. "
        f"The discussion maintains a {row.tone} tone while outlining key events and perspectives. "
        f"Readers can expect insights into innovations, challenges, and wider implications across the field of {row.topic.lower()}. "
        f"It offers a concise overview meant to inform readers about the continuing evolution in this domain."
    )

tqdm.pandas()
df["content"] = df.progress_apply(synthesize, axis=1)

df.to_csv("feeds_rows_enriched.csv", index=False)
print("✅ Enriched dataset saved as feeds_rows_enriched.csv")
```
